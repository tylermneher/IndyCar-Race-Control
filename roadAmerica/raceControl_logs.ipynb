```python
import pandas as pd

list = []
f = open('racecontrol_indycar_com_1655084049571.log')
line = f.readline()

while line:
    try:
        list.append(line)
    except ValueError:
        print('Error in line :' + line )
    line = f.readline()
```

```python
for i in list:
    print(i)
```

```python
df = pd.DataFrame([sub.split(" : ") for sub in list])
df
```
